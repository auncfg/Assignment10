### 문제 05
**모든 정점 사이에 간선이 존재하는 그래프 — 인접 행렬 vs 인접 리스트**

**→ 인접 행렬(Adjacency Matrix) 선택**

**이유:**
- 모든 정점 사이에 간선이 존재하는 완전 그래프(Complete Graph)는 정점 수 n에 대해 간선 수가 n(n-1)/2 (무방향) 또는 n(n-1) (방향)으로 매우 밀집되어 있음
- 인접 리스트는 간선이 적을 때 메모리를 절약하는 구조인데, 완전 그래프에서는 어차피 모든 간선을 저장해야 하므로 포인터 오버헤드만 추가됨
- 인접 행렬은 두 정점 u, v 사이의 간선 존재 여부를 **O(1)**에 확인 가능
- 완전 그래프에서 인접 행렬의 메모리 O(n²)은 간선 수 O(n²)와 동일하므로 낭비가 없음

### 문제 06
**희소 그래프(Sparse Graph) — 인접 행렬 vs 인접 리스트**

**→ 인접 리스트(Adjacency List) 선택**

**이유:**
- 희소 그래프는 간선 수 E가 정점 수 V에 비해 매우 적음 (E ≪ V²)
- 인접 행렬은 간선 존재 여부와 무관하게 항상 O(V²) 메모리를 사용 → 대부분의 칸이 0으로 낭비
- 인접 리스트는 실제 존재하는 간선만 저장하므로 O(V + E) 메모리 사용 → 메모리 효율적
- DFS, BFS 등 탐색 시 실제 이웃만 순회하므로 시간도 절약됨

### 문제 08
**DFS(1) 수행 시 v=1인 상태에서 DFS(x) 호출 횟수**

```
DFS(v):
    v.visited ← YES
    for each x ∈ v.adj
        if (x.visited = NO)
            (1) DFS(x)
```

**풀이 원리:**
- DFS(1) 호출 시 v=1 상태에서 `for each x ∈ v.adj` 루프를 돔
- 이 중 **x.visited = NO인 x의 수**만큼 DFS(x)가 호출됨
- 즉, **정점 1의 인접 정점 중 아직 방문되지 않은 정점의 수**가 답

**→ 정점 1의 인접 정점 수만큼 호출됨 (모두 미방문 상태이므로)**

### 문제 16
**위상 정렬 (STACK 사용)**

그래프 간선: a→d, a→b, b→c, d→e, d→f, e→f, e→g, f→g, f→h

#### 진입 차수(in-degree) 계산

| 정점 | 진입 간선       | in-degree |
|------|-----------------|----------|
| a    | 없음            | 0        |
| b    | a→b             | 1        |
| c    | b→c             | 1        |
| d    | a→d             | 1        |
| e    | d→e             | 1        |
| f    | d→f, e→f        | 2        |
| g    | e→g, f→g        | 2        |
| h    | f→h             | 1        |

#### STACK을 사용한 위상 정렬 수도 코드

```
TopologicalSort(G):
    각 정점 v의 in-degree 계산

    STACK S
    for each v in G:
        if v.inDegree = 0:
            S.push(v)

    result ← []
    while S is not empty:
        v ← S.pop()
        result.append(v)
        for each x ∈ v.adj:
            x.inDegree ← x.inDegree - 1
            if x.inDegree = 0:
                S.push(x)

    return result
```

#### 수행 과정

초기: in-degree=0 → **a** push → Stack: [a]

| Step | pop | result         | 변화                                  | Stack   |
|------|-----|----------------|---------------------------------------|---------|
| 1    | a   | [a]            | b: 1→0 push, d: 1→0 push             | [d, b]  |
| 2    | b   | [a, b]         | c: 1→0 push                           | [d, c]  |
| 3    | c   | [a, b, c]      | 나가는 간선 없음                      | [d]     |
| 4    | d   | [a, b, c, d]   | e: 1→0 push, f: 2→1                  | [e]     |
| 5    | e   | [a, b, c, d, e]| f: 1→0 push, g: 2→1                  | [f]     |
| 6    | f   | [a,b,c,d,e,f]  | g: 1→0 push, h: 1→0 push            | [g, h]  |
| 7    | h   | [a,b,c,d,e,f,h]| 나가는 간선 없음                      | [g]     |
| 8    | g   | [a,b,c,d,e,f,h,g]| 나가는 간선 없음                    | []      |

**결과: a → b → c → d → e → f → h → g**

#### 가능한 위상 정렬 순서의 수

- a는 항상 첫 번째 (유일한 in-degree 0)
- b체인(b→c)과 d체인(d→e→f→...)은 독립적으로 섞일 수 있음
- c는 {d 전, d~e 사이, e~f 사이, f~g,h 사이, g,h 이후} 중 위치 가능 → **6가지**
- g와 h의 순서 → **2가지**

**총 가능한 위상 정렬 순서: 6 × 2 = 12가지**

  A ---(+5)---> B

### 문제 22
**벨만-포드 알고리즘에서 답을 구할 수 없는 경우**

**→ 그래프에 음의 사이클(Negative Cycle)이 존재할 때**

**이유:**
- 벨만-포드는 음의 가중치 간선은 처리 가능
- 그러나 음의 사이클이 존재하면, 그 사이클을 반복할수록 경로 비용이 무한히 감소
- 최단 거리가 -∞로 발산하여 수렴하지 않음
- 벨만-포드는 V-1번 완화(relaxation) 후 한 번 더 시도하여 거리가 줄어드는 정점이 있으면 음의 사이클 존재로 판단 → 답 없음 반환

```
예시: A→B(1), B→C(-3), C→A(1)
사이클 합: 1 + (-3) + 1 = -1 < 0 → 음의 사이클
→ 벨만-포드 답 구할 수 없음
```

#### (3) 퀵 정렬 (Quick Sort)

```

→ 좌: []  |  [12]  |  우: [70, 30, 20, 55, 25, 40, 50]

[70, 30, 20, 55, 25, 40, 50]
→ 좌: [30, 20, 55, 25, 40, 50]  |  [70]  |  우: []

3단계: pivot = 30
[30, 20, 55, 25, 40, 50]
→ 좌: [20, 25]  |  [30]  |  우: [55, 25, 40, 50]

   [20, 25] → pivot = 20
   → 좌: []  |  [20]  |  우: [25]

   [55, 40, 50] → pivot = 55
   → 좌: [40, 50]  |  [55]  |  우: []

      → 좌: []  |  [40]  |  우: [50]
      → [40, 50]
   → [40, 50, 55]

```

**최종: [12, 20, 25, 30, 40, 50, 55, 70]**

````python
# ============================================================
# 정렬 알고리즘 검증 코드
# ============================================================

def selection_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

def bubble_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                swapped = True
        if not swapped:
            break
    return a

def insertion_sort(arr):
    a = arr[:]
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
    return a

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1
    return result + left[i:] + right[j:]

def heapify(arr, n, i):
    largest = i
    l, r = 2*i+1, 2*i+2
    if l < n and arr[l] > arr[largest]: largest = l
    if r < n and arr[r] > arr[largest]: largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n//2-1, -1, -1):
        heapify(a, n, i)
    for i in range(n-1, 0, -1):
        a[0], a[i] = a[i], a[0]
        heapify(a, i, 0)
    return a

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]
    left  = [x for x in arr[1:] if x <= pivot]
    right = [x for x in arr[1:] if x >  pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)

# ---- 검증 ----
arr1 = [12, 70, 30, 20, 55, 25]
arr2 = [12, 70, 30, 20, 55, 25, 40, 50]

print("=== 문제 05 ===")
print("선택 정렬:", selection_sort(arr1))
print("버블 정렬:", bubble_sort(arr1))
print("삽입 정렬:", insertion_sort(arr1))

print("\n=== 문제 06 ===")
print("병합 정렬:", merge_sort(arr2))
print("힙   정렬:", heap_sort(arr2))
print("퀵   정렬:", quick_sort(arr2))
````
